## 1D Plottinging


### 1.Histogram

```python
data = np.random.randn(1000)  # Generate random data

print(data)
plt.hist(data, bins=100, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.grid(True)
plt.show()
```

### BarChart

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 15, 25, 18]

plt.bar(categories, values, color='orange')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

## 2D Plotting

### 1.Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', label='Data Points')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot Example')
plt.legend()
plt.show()
```

### 2.Line Plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', label='Sine Wave')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Line Plot Example')
plt.legend()
plt.grid(True)
plt.show()
```

### 3.Contour Plot

```python
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Contour Plot Example')
plt.colorbar()
plt.show()
```

### 4.Heatmap

```python
data = np.random.rand(10, 10)

sns.heatmap(data, cmap='coolwarm', annot=True)
plt.title('Heatmap Example')
plt.show()
```

## 3D Plotting

### 1.Scatter Plots

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter(x, y, z, color='green')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.title('3D Scatter Plot Example')
plt.show()
```

### 2.Surfface Plot

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

ax.plot_surface(X, Y, Z, cmap='coolwarm')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')
plt.title('3D Surface Plot Example')
plt.show()
```

## Higher Dimensional Plots

### 1. Parallel Coordinatestes Plot

```python
from pandas.plotting import parallel_coordinates
import pandas as pd

df = pd.DataFrame({
    'Feature1': np.random.rand(10),
    'Feature2': np.random.rand(10),
    'Feature3': np.random.rand(10),
    'Class': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
})

parallel_coordinates(df, 'Class', colormap=plt.get_cmap("Set1"))
plt.title('Parallel Coordinates Plot Example')
plt.show()
```

### 2.Andrew Plots

```python
from pandas.plotting import andrews_curves

andrews_curves(df, 'Class', colormap='coolwarm')
plt.title('Andrews Plot Example')
plt.show()
```
